```python
class Human:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def check_adult(self):

        if self.age >= 20:
            print (self.name + "は大人です。")
        else:
            print (self.name + "は大人でない。")

Humans = [
         Human("テスト1",18),
         Human("テスト2",19),
         Human("テスト3",20),
         Human("テスト4",21),
         Human("テスト5",22),
]

for human in Humans:
 human.check_adult()
```
